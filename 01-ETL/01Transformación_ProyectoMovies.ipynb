```python
import pandas as pd

data = pd.read_csv("/content/movies_dataset.csv")
print(data.shape)  # Verificar la cantidad de filas y columnas
print(data.columns)  # Obtener los nombres de las columnas
print(data.head())  # Mostrar las primeras filas del dataset
```

```python

# Desanidar la columna 'belongs_to_collection' y obtener el nombre
def extract_collection_name(x):
    try:
        collection = ast.literal_eval(x)
        return collection['name']
    except (ValueError, TypeError, KeyError):
        return np.nan

data['belongs_to_collection'] = data['belongs_to_collection'].apply(extract_collection_name)

# Verificar los resultados
print(data['belongs_to_collection'])

```

```python

#Desanidar la columna 'genres' y obtener el nombre
# Función de extracción de nombres de género
def extract_genre_names(x):
    try:
        genres = eval(x)
        genre_names = [genre['name'] for genre in genres]
        return ', '.join(genre_names)
    except:
        return ''

# Desanidar la columna 'genres' y obtener los nombres como una cadena
data['genres'] = data['genres'].apply(lambda x: extract_genre_names(x))

# Verificar los resultados
print(data['genres'])




```

```python

#Desanidar la columna "production_companies" y obtener el nombre
# Función de extracción de nombres de production_companies:
import ast
import pandas as pd
import numpy as np

# Función de extracción de nombres de compañías de producción
def extract_company_names(x):
    try:
        companies = ast.literal_eval(x)
        company_names = [company['name'] for company in companies]
        return ', '.join(company_names)
    except:
        return ''

# Desanidar la columna 'production_companies' y obtener los nombres como una cadena
data['production_companies'] = data['production_companies'].apply(lambda x: extract_company_names(x))

# Verificar los resultados
print(data['production_companies'])

```

```python
#production_countries,spoken_languages 
#Desanidar la columna "production_countries" y obtener el nombre
import ast
import pandas as pd
import numpy as np

# Función de extracción de nombres de países de producción
def extract_country_names(x):
    try:
        countries = ast.literal_eval(x)
        country_names = [country['name'] for country in countries]
        return ', '.join(country_names)
    except:
        return ''

# Desanidar la columna 'production_countries' y obtener los nombres como una cadena
data['production_countries'] = data['production_countries'].apply(lambda x: extract_country_names(x))

# Verificar los resultados

print(data['production_countries'].head(10))

```

```python
#Desanidar la columna "spoken_languages " y obtener el nombre
import ast
import pandas as pd
import numpy as np

# Función de extracción de nombres de idiomas hablados
def extract_language_names(x):
    try:
        languages = ast.literal_eval(x)
        language_names = [language['name'] for language in languages]
        return ', '.join(language_names)
    except:
        return ''

# Desanidar la columna 'spoken_languages' y obtener los nombres como una cadena
data['spoken_languages'] = data['spoken_languages'].apply(lambda x: extract_language_names(x))

# Verificar los resultados
print(data['production_countries'].head(50))
```

```python
#Los valores nulos de los columnas " revenue y  budget"deben ser rellenados por el número 0.
# Rellenar los valores nulos en las columnas 'revenue' y 'budget' con 0
data['revenue'] = data['revenue'].fillna(0)
data['budget'] = data['budget'].fillna(0)

# Verificar los resultados
print(data[['revenue', 'budget']])
```

```python
# Eliminar los valores nulos en la columna 'release_date'
data = data.dropna(subset=['release_date'])

# Verificar los resultados
print(data['release_date'])
```

```python
#La columna : "release_date", deberán tener el formato AAAA-mm-dd
import pandas as pd

# Convertir la columna 'release_date' a tipo datetime
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Formatear la columna 'release_date' en el formato AAAA-mm-dd
data['release_date'] = data['release_date'].dt.strftime('%Y-%m-%d')

# Verificar los resultados
print(data['release_date'])
```

```python
#crear la columna release_year donde extraerán el año de la columna :"release_date"
import pandas as pd

# Convertir la columna 'release_date' a tipo datetime
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Extraer el año y crear la columna 'release_year'
data['release_year'] = data['release_date'].dt.year
data['release_year'] = data['release_year'].fillna(0).astype(int)

# Verificar los resultados
print(data['release_year'])

```

```python
#
# Eliminar las columnas no utilizadas
columns_to_drop = ['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage']
data = data.drop(columns=columns_to_drop)

# Verificar los resultados
print(data.head())
```

```python
#Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, 
#dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0
# Función para calcular el retorno de inversión

# Función para calcular el retorno de inversión
def calculate_return(row):
    revenue = row['revenue']
    budget = row['budget']
    if pd.notnull(revenue) and pd.notnull(budget):
        try:
            revenue = float(revenue)
            budget = float(budget)
            if budget != 0:
                return revenue / budget
            else:
                return 0
        except ValueError:
            return 0
    else:
        return 0

# Crear la columna 'return' con el retorno de inversión
data['return'] = data.apply(calculate_return, axis=1)

# Verificar los resultados
print(data['return'])

```
